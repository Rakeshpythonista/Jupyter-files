### Part-1.....Workbook created of name 'IPL_2020_UAE'

```python
Team = pd.read_html('https://www.icccricketschedule.com/vivo-ipl-2020-schedule-team-venue-time-table-pdf-point-table-ranking-winning-prediction/')
```

```python
TD = Team[0]
TD.head()
```

```python
with pd.ExcelWriter('IPL_2020_UAE.xlsx') as writer:
    TD.to_excel(writer, sheet_name='Schedule', index=False)
```

### Part-2.....To add multiple excel sheet  to Workbook of name 'IPL MVP Stats'

```python
data = pd.read_html('https://sportspages.in/cricket/ipl/ipl-stat/')
```

```python
df1 = data[0]
df1.head()
```

```python
df2 = data[1]
df2.head()
```

```python
df3 = data[2]
df3.head()
```

```python
with pd.ExcelWriter('IPL_MVP_Stats.xlsx') as writer:
    df1.to_excel(writer, sheet_name='MVP Player', index=False)
    df2.to_excel(writer, sheet_name='Batting Statistics', index=False)
    df3.to_excel(writer, sheet_name='Bowling Statistics', index=False)
```

```python
check = pd.read_excel('IPL_MVP_Stats.xlsx', sheet_name=['MVP Player','Batting Statistics','Bowling Statistics'])
```

```python
check['MVP Player'].head(3)
```

```python
check['Batting Statistics'].head(3)
```

```python
check['Bowling Statistics'].head(3)
```

### Part-3.... Append to an existing excel workbook

```python
data = pd.read_html('https://www.iplt20.com/stats/2019/player-points')
```

```python
New_data = data[0]
New_data.head()
```

```python
with pd.ExcelWriter('IPL_MVP_Stats.xlsx', engine='openpyxl', mode='a') as writer:
    New_data.to_excel(writer, sheet_name='player-points', index=False)
```
